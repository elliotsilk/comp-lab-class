```python
# this notebook is for plotting
# the end to end analysis
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
```

```python
import os
# change python directory to /home/gmh4/projects/CompLab/comp-lab-class

# load .xvg file including "@" and "#" in the list
time, PE, KE, totE, temp, pres, vol = np.loadtxt("../Data/1hz3_T310.run.25000000.energy.xvg", comments = ["@", "#"], unpack=True)
```

```python
# to use function, n = the variable you want to plot; var = string title to label y axis and title
# PE = potential energy (kJ/mol)
# KE = kinetic energy (kJ/mol)
# totE = total energy (kJ/mol)
# temp = temperature (K)
# pres = pressure (bar) 
# vol = volume (nm^3)
def timeVsVar(n, var):
    time, PE, KE, totE, temp, pres, vol = np.loadtxt("../Data/1hz3_T310.run.25000000.energy.xvg", comments = ["@", "#"], unpack=True)
    timeps = []
    for x in time:
        timeps.append(x/1000)
    x = timeps
    y = n
    plt.xlabel('Time (ps)')
    plt.ylabel(var)
    plt.title('Time (ps) Versus ' + var, fontsize=20)
    plt.plot(x, y)
```

```python
timeVsVar(PE, 'Potential Energy (kJ/mol)')
```

```python
fig, ax1 = plt.subplots(figsize=(8, 8))
ax2 = ax1.twinx()

ax1.plot(time, PE, color='k')
ax2.plot(time, KE, color='g')

ax1.set_xlabel("Time (ps)")
ax1.set_ylabel("Potential Energy (kJ/mol)")
ax2.set_ylabel("Kinetic Energy (kJ/mol)")
fig.suptitle("Potential Energy and Kinetic Energy Versus Time", fontsize=20)
```

```python
ub = md.load_pdb("../Data/1UBQ_processed.pdb")

hbonds = md.baker_hubbard(ub, periodic=False)
label = lambda hbond : '%s -- %s' % (ub.topology.atom(hbond[0]), ub.topology.atom(hbond[2]))
for hbond in hbonds:
    print(label(hbond))

#counts h bonds
hel = md.compute_dssp(ub)
size = np.size(hel)
i = 0
count = 0

while i < size:
    if hel [0][i] == 'H': 
        count += 1
        i += 1
    else:
        i += 1
    
print (count)
```
