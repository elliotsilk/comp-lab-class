```python
# this notebook is for plotting
# the end to end analysis
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
import os
```

```python
# use numpy to load COLVAR data
os.chdir("/home/ess545/comp-lab-class/Week6/")
#plot using np
colvarA = np.genfromtxt('exercise1/topolA/COLVAR', skip_header=1)
colvarB = np.genfromtxt('exercise1/topolB/COLVAR', skip_header=1)

phi_A_c = colvarA[:,1]
psi_A_c = colvarA[:,2]
phi_B_c = colvarB[:,1]
psi_B_c = colvarB[:,2]

#use traj to compute phi and psi
trajA = md.load ("exercise1/topolA/traj_comp.xtc", top='exercise1/topolA/confout.gro')
trajB = md.load ("exercise1/topolB/traj_comp.xtc", top='exercise1/topolB/confout.gro')
psi_A = md.compute_psi (trajA)
phi_A = md.compute_phi (trajA)
psi_B = md.compute_psi (trajB)
phi_B = md.compute_phi (trajB)

#plot phi_A_c and phi_A
plt.title("COLVAR Phi A versus .xtc output Phi A")
plt.xlabel("Time")
plt.ylabel("Radians")
plt.plot(phi_A_c, label='phi_A_c')
plt.plot(phi_A[1], label='phi_A')
plt.legend()
plt.show()

#plot psi_A_c and psi_A
plt.title("COLVAR Psi A versus .xtc output Psi A")
plt.xlabel("Time")
plt.ylabel("Radians")
plt.plot(psi_A_c, label='psi_A_c')
plt.plot(psi_A[1], label='psi_A')
plt.legend()
plt.show()

#plot phi_B_c and phi_B
plt.title("COLVAR Phi B versus .xtc output Phi B")
plt.xlabel("Time")
plt.ylabel("Radians")
plt.plot(phi_B_c, label='phi_B_c')
plt.plot(phi_B[1], label='phi_B')
plt.legend()
plt.show()

#plot psi_B_c and psi_B
plt.title("COLVAR Psi B versus .xtc output Psi B")
plt.xlabel("Time")
plt.ylabel("Radians")
plt.plot(psi_B_c, label='psi_B_c')
plt.plot(psi_B[1], label='psi_B')
plt.legend()
plt.show()

# compute standard deviation of phi_A_c and psi_A_c
standard_deviation_phi_A_c = np.std(phi_A_c)
standard_deviation_psi_A_c = np.std(psi_A_c)
print(standard_deviation_phi_A_c)
print(standard_deviation_psi_A_c)
# compute standard deviation of phi_B_c and psi_B_c
standard_deviation_phi_B_c = np.std(phi_B_c)
standard_deviation_psi_B_c = np.std(psi_B_c)
print(standard_deviation_phi_B_c)
print(standard_deviation_psi_B_c)

```
