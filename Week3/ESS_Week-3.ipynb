```python
# install library
import matplotlib.pyplot as plt
import numpy as np
import os

# graph pressure, density, volume from equilibration; RMS, rate of gyration; energy min

# set up path
os.chdir ("/home/ess545/comp-lab-class/Week3/Data")

# load all data
x1, y1 = np.loadtxt("potential.xvg", comments=["@", "#"], unpack=True)
x2, y2 = np.loadtxt("density.xvg", comments=["@", "#"], unpack=True)
x3, y3 = np.loadtxt("pressure.xvg", comments=["@", "#"], unpack=True)
x4, y4 = np.loadtxt("temperature.xvg", comments=["@", "#"], unpack=True)
# x5, y5 = np.loadtxt("gyrate.xvg", comments=["#"], unpack=True)
x6, y6 = np.loadtxt("rmsd.xvg", comments=["@", "#"], unpack=True)

#plot potential
plt.figure()
plt.plot(x1, y1)
plt.xlabel("Production Step")
plt.ylabel("Potential (kJ/mol)")
plt.title("Potential Energy")
plt.savefig("potential.png")
plt.show()

#plot density
plt.figure()
plt.plot(x2, y2)
plt.xlabel("Time (ps)")
plt.ylabel("Density (g/cm^3)")
plt.title("Density")
plt.savefig("density.png")
plt.show()

#plot pressure
plt.figure()
plt.plot(x3, y3)
plt.xlabel("Time (ps)")
plt.ylabel("Pressure (bar)")
plt.title("Pressure")
plt.savefig("pressure.png")
plt.show()

#plot temperature
plt.figure()
plt.plot(x4, y4)
plt.xlabel("Time (ps)")
plt.ylabel("Temperature (K)")
plt.title("Temperature")
plt.savefig("temperature.png")
plt.show()

#plot RMSD
plt.figure()
plt.plot(x6, y6)
plt.xlabel("Time (ps)")
plt.ylabel("RMSD (nm)")
plt.title("RMSD")
plt.savefig("rmsd.png")
plt.show()

```
